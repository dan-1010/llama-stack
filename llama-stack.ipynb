How to setup linux in windows

```python
Enable virtualisation in bios.
Enable Windows Subsytem for linux in Turn Windows features or or off setting.
# 1. Install Windows Terminal and WSL from microsoft store if not available.
# 2. "wsl --update" command to update to the latest version of Windows Subsytem For Linux.
wsl --set-default-version 2 to set it to the latest version
# 3. 'wsl --list --online>' use the coommand to get the version of linux. 
# 4. 'wsl --install -d Ubuntu-22.04' command to install the desired version.
# 5. Search for Ubuntu app installed and run it as admin
# 6. Set username and password 


https://www.youtube.com/watch?v=AMlaEFaKG88 link to setup wsl in windows.
```
